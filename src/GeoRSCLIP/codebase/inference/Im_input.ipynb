```python
import tifffile
import torchvision.transforms as T
from torchvision.datasets import ImageFolder
```

```python
class BuildingLoader(ImageFolder):
    def __init__(self, root, 
                 transform=T.Compose([T.ToTensor()])):
        super().__init__(root,
                       transform=transform, 
                       loader=tifffile.imread, )
        
```
